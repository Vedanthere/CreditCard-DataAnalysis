## Analyses
I will explore and compare the relationships between credit limit, average utilization ratio, age, and gender on a credit card with in detail analysis, consisting of in depth visualizations.

### Distributions
Firstly, I will start exploring distributions in our Credit Card data. Figure 1, below shows the distribution of credit limit on the credit card. As we can see, the majority of customers have a credit limit of $10,000 and below.

```python
#Credit limit  (Figure 1)   
plt.title('Credit Limit Distribution')
plt.xlabel('Credit Limit')
plt.ylabel('Number of Customers')
creditlimit = dataset.Credit_Limit.hist(bins=7, figsize=(7,5))
creditlimit
```

Below plot (Fig 3.) shows that the most of credit card customers are Female.

```python
#Gender (Figure 2)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
gender=dataset.Gender.hist(bins=4, figsize=(7,5))
gender
#M for Male
#F for Female
```

Below plot (Fig 4.) shows that an average utilization ratio is mostly between 0.0 and 0.3. So, it looks like they have an excellent credit score.

```python
#Utilization Ratio (Figure 3)
plt.title('Average Utilization Ratio Distribution')
plt.xlabel('Average Utilization Ratio')
plt.ylabel('Number of Customers')
utilizationRatio = dataset.Avg_Utilization_Ratio.hist(bins=4, figsize=(7,5))
utilizationRatio
```

```python
#Min, Mean, Max of Average of Utilization Ratio
ds = dataset.Avg_Utilization_Ratio.describe()
print(f"Minimum Utlilization Ratio: {round(ds['min'],2)}")
print(f"Mean Utlilization Ratio: {round(ds['mean'],2)}")
print(f"Maximum Utlilization Ratio: {round(ds['max'],2)}")
```

As, we can see that the mean Utilization Ratio is 0.3 same as we predicted on the graph (Figure 4).

### Credit Limit and Average Utilization Ratio

This is the relationship between Credit Limit and Average Utilization Ratio of Credit Card data.

```python
#Figure 4
AvgPlot = sns.relplot(data=dataset, x=dataset['Avg_Utilization_Ratio'], y=dataset['Credit_Limit'])
```

As per the above plot graph, increasing the credit limit lowers the credit utilization ratio and vice versa (the higher an average utilization ratio lowers the credit limit). Due to high utilization ratio is a sign that customers may be experiencing financial difficulty and is a strong indicator of lending risk so, it forces them to keep lower the credit limit.

### Credit Limit and Gender

An examination of this table, calculates the number of females and males that have the same amount of Credit Limit.

```python
# Grouped operations
group_data = dataset.groupby(['Credit_Limit','Gender']).size().unstack()
group_data
```

I can explore the data more in depth using the dropna function to remove all rows with Nan values from an interaction of Credit Limit and Gender table.

```python
#dropna function
group_data_2 = group_data.dropna(axis=0)
group_data_2
```

I have further modified this table, now we only have a total of 724 rows. So, interaction between Credit Limit and Gender had 4,751 instances of missing data (Null).

```python
gender_creditLimit = dataset[['Gender', 'Credit_Limit']]  
gender_creditLimit
```

Now, I am plotting the overview graph for correlation between Credit Limit and Gender.

```python
#Figure 5
plt.figure(figsize = (15,8))

plt.hist([gender_creditLimit.loc[gender_creditLimit['Gender'] == 'M']['Credit_Limit'].tolist(), gender_creditLimit.loc[gender_creditLimit['Gender'] == 'F']['Credit_Limit'].tolist()], bins = 10,color = ['blue','red'],
rwidth = 0.9)
plt.legend(['Male','Female'])
plt.xlabel('Credit Limit ($)')
plt.ylabel('Frequency between Male and Female')
title = "Credit Limit by Gender"
plt.title(title)
plt.show() 
```

As this plot graph shows, the Female (red bar plot) has less credit limit than Male (blue bar plot) in a range of between $10,000 and $35,000. And, females have more credit limits than Male within a range of $5,000 and $10,000. From $0 to $5,000 there are approximately 1,800 females that have more credit limits than the males.

### Average Utilization Ratio and Gender

```python
#Figure 6
colors = ['#78C850', '#7038F8']
plt.figure(figsize=(15,6))
sns.violinplot(x='Gender', y='Avg_Utilization_Ratio', data=dataset, 
               palette=colors) 
```

As this plot indicates that the mean of male average utilization ratio is about 0.1 and the mean of female average utilization ratio is 0.4. Also, the number of male average utilization ratio is wider at about 0.1 and the number of female utilization ratio is quite wider in a range between 0.6 and 0.8. This shows that females have more utilization ratio than males.
